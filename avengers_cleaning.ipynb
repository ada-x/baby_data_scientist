```python
import matplotlib.pyplot as plt
%matplotlib inline

avengers['Year'].hist()
```

```python
# filter out all pre-1960 avengers entries

true_avengers = avengers[avengers['Year'] >= 1960]
true_avengers
```

### now to consolidate the separate deaths columns into one
### add to true_avengers dataframe

```python
columns = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
# to get an idea of what it looks like
true_avengers[columns] 
```

```python
def clean_deaths(row):
    num_deaths = 0
    columns = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
    
    for c in columns:
        death = row[c]
        
        if pd.isnull(death) or death == 'NO':
            continue
        elif death == 'YES':
            num_deaths += 1
    
    return num_deaths
    
true_avengers.loc['Deaths'] = true_avengers.apply(lambda row: clean_deaths(row), axis=1)
true_avengers
```

### last section: check the accuracy of the 'years since joining' values
### by subtracting the current year (this project used 2015 as 'current') from the year the avenger was introduced
### and comparing that with the 'years since joining' value

```python
# make sure we output an integer
accurate_year_count = int()
correct_years_list = true_avengers[true_avengers['Years since joining'] == (2015 - true_avengers['Year'])]
accurate_year_count = len(correct_years_list)

accurate_year_count
```
